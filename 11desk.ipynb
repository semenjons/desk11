Нахождение корней с помощью бисекции

```python
import numpy as np
```

```python
def bisection(f,lhs,rhs,eps=1e-6):
    mid = (lhs+rhs)/2
    
    while abs(lhs-rhs) > eps:
        mid = (lhs+rhs)/2
        
        if f(lhs) * f(mid) > 0:
            lhs =mid
        else:
            rhs =mid
    return mid
```

```python
def func(x):
    return -12 * x**4 * np.sin(np.cos(x))-18 * x**3 + 5 * x**2 + 10 * x - 30
```

Корень в диапазоне  {-8, -7}

```python
import numpy as np
lhs = -8
rhs = -7
N = 100
mid = (lhs+rhs)/2
x = np.linspace(lhs,rhs,N)
y = [func(x_i) for x_i in x]
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y);
```

```python
bisection(func, -8, -7)
```

Производная {-8, -7}

```python
def diff (f,x,h):
    N = x.shape[0]    
    d = np.zeros(N,dtype= np.float32)
    d[0]=(-3*f(x[0])+4*f(x[1])-f(x[2]))/(2*h)
    for i in range(1,N-1):
        d[i] = (f(x[i+1])-f(x[i-1]))/(2*h)
    d[N-1]=(f(x[N-3])-4*f(x[N-2])+3*f(x[N-1]))/(2*h)
    return d
    
```

```python
h = (rhs - lhs)/(N-1)
dif = diff(func, x, h)
```

```python
plt.plot(x,dif);
```
